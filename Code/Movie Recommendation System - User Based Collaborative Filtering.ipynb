## Popularity Based Recommendation System

```python
num_ratings = final_df.groupby('title').count()['rating'].reset_index()
```

```python
num_ratings.rename(columns={'rating':'num_ratings'},inplace=True)
```

```python
num_ratings
```

```python
avg_ratings = final_df.groupby('title').mean()['rating'].reset_index()
avg_ratings.rename(columns={'rating':'average_rating'},inplace=True)
avg_ratings
```

```python
popular_df = num_ratings.merge(avg_ratings,on='title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>100].sort_values('average_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
x = final_df.groupby('userId').count()['rating'] > 100
```

```python
users_with_200_plus_ratings = x[x].index
```

```python
df = final_df[final_df['userId'].isin(users_with_200_plus_ratings)]
```

```python
df.shape
```

```python
y = df.groupby('title').count()['userId']>20
```

```python
movies_with_25_ratings = y[y].index
```

```python
df_final = df[df['title'].isin(movies_with_25_ratings)]
```

```python
df_final.shape
```

```python
df_final
```

```python
pivot = df_final.pivot_table(columns='userId',index='title',values='rating')
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot)
```

```python
similarity_scores.shape
```

```python
def recommend_movie(movie_name):
    index = np.where(pivot.index == movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df_final[df['title'].reindex(df_final.index) == pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['genres'].values))
        item.extend(list(temp_df.drop_duplicates('title')['rating'].values))
        item.extend(list(temp_df.drop_duplicates('title')['movieId'].values))
        data.append(item)
    return data
```

```python
recommend_movie('Zombieland')
```
